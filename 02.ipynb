Question 1: What is a random variable in probability theory?
 - A random variable in probability theory is a variable that represents the possible numerical outcomes of a random experiment. Its value depends on chance and is determined by the outcome of the experiment. For example, when a coin is tossed, a random variable can be defined as the number of heads obtained, which can take values 0 or 1. Random variables are used to quantify uncertain events and are usually classified as discrete (taking countable values) or continuous (taking values within a range).

 Question 2: What are the types of random variables?

Continuous random variables ‚Äì These take any value within a given range. They usually arise from measurements.

Question 3: Explain the difference between discrete and continuous distributions.
 - The difference between discrete and continuous distributions lies in the nature of the values their random variables can take. A discrete distribution deals with discrete random variables that take countable and specific values, such as 0, 1, 2, and so on. Each possible value has a definite probability associated with it. Examples include the number of students in a class or the number of heads obtained in coin tosses. In contrast, a continuous distribution deals with continuous random variables that can take any value within a given range. The probability is defined over an interval rather than at a single point. Examples include height, weight, or time. Continuous distributions are usually represented using probability density functions rather than individual probabilities.

 Question 4: What is a binomial distribution, and how is it used in probability?
  - A binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials of a random experiment. Each trial has only two possible outcomes, usually called success and failure, and the probability of success remains the same in every trial. The binomial distribution is defined by two parameters: the number of trials ùëõ and the probability of success ùëù.
It is used in probability to model situations such as the number of heads in repeated coin tosses, the number of defective items in a batch, or the number of students who pass an exam. By using the binomial distribution, we can calculate the probability of getting a specific number of successes in a given number of trials.

Question 7: What is the significance of confidence intervals in statistical analysis?
 - Confidence intervals (CIs) are an important concept in statistical analysis because they provide a range of values within which the true population parameter (such as the mean or proportion) is likely to lie with a specified level of confidence, usually expressed as 90%, 95%, or 99%.

The significance of confidence intervals is that they not only give an estimate of the parameter but also indicate the uncertainty or reliability of that estimate. For example, saying that the average height of students is 165 cm with a 95% confidence interval of 162‚Äì168 cm means we are 95% confident that the true average height of all students lies between 162 and 168 cm. Confidence intervals are widely used in research, business, and science to make informed decisions and draw conclusions from sample data without measuring the entire population.


Question 8: What is the concept of expected value in a probability distribution?
 - The expected value of a probability distribution is the long-term average or mean value that a random variable is likely to take if an experiment is repeated many times. It represents the ‚Äúcenter‚Äù or balance point of the distribution and gives an idea of the typical outcome. For a discrete random variable, the expected value is calculated by multiplying each possible value by its probability and then summing these products. For example, when rolling a fair six-sided die, the expected value is 3.5, meaning that over many rolls, the average outcome will tend toward 3.5. The concept of expected value is important in decision-making, economics, insurance, and gambling, as it helps predict average outcomes and make informed choices under uncertainty.




```python
# Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
# deviation using NumPy, and draw a histogram to visualize the distribution.
#  - Code
# Python program to generate random numbers from a normal distribution
# and compute mean, standard deviation, and plot a histogram

import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean = 50
std_dev = 5
n = 1000

# Generate 1000 random numbers from normal distribution
data = np.random.normal(mean, std_dev, n)

# Compute mean and standard deviation
calculated_mean = np.mean(data)
calculated_std = np.std(data)

print("Calculated Mean:", calculated_mean)
print("Calculated Standard Deviation:", calculated_std)

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Output
Calculated Mean: 49.97
Calculated Standard Deviation: 5.13
```

```python
# Question 10: You are working as a data analyst for a retail company. The company has
# collected daily sales data for 2 years and wants you to identify the overall sales trend.
# daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
#  235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
# ‚óè Explain how you would apply the Central Limit Theorem to estimate the average sales
# with a 95% confidence interval.
# ‚óè Write the Python code to compute the mean sales and its confidence interval.

 - To identify the overall sales trend, we can use the Central Limit Theorem (CLT).
Even if the daily sales data is not perfectly normally distributed, the CLT
tells us that the sampling distribution of the sample mean will approximate a
normal distribution if we take a sufficiently large number of samples. This
allows us to estimate the average daily sales and calculate a 95% confidence
interval, which gives a range where we are 95% confident that the true mean
daily sales of the company lies.

# Code
# Python code to compute mean sales and 95% confidence interval

import numpy as np
from scipy import stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Sample size
n = len(daily_sales)

# Sample mean and standard deviation
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # sample standard deviation

# 95% confidence interval using t-distribution
confidence_level = 0.95
t_value = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
margin_of_error = t_value * (std_sales / np.sqrt(n))

lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```
